```python
a = 32626816
target_range = (40000000, 40001000)

for x_int in range(1, 601):  # x를 1부터 600까지 변화시킵니다. (0.01 * 100 * 6 = 6)
    for y_int in range(1, 1001):  # y를 1부터 1000까지 변화시킵니다. (0.01 * 100 * 10 = 10)
        x = round(x_int / 100, 2)  # x를 소수점 두 자리까지 변환합니다.
        y = round(y_int / 100, 2)  # y를 소수점 두 자리까지 변환합니다.

        b = 0.01 * x * a
        c = 0.01 * y * (a + b)
        d = a + b + c
        e = 0.1 * d

        if target_range[0] <= (d + e) <= target_range[1]:  # 오차 범위 내에서 d + e가 target_sum과 일치하는지 확인합니다.
            print(f"x: {x:.2f}, y: {y:.2f}")
            print(f"a: {a}, b: {b}, c: {c}, d: {d}, e: {e}")
            print(f"d + e: {d + e}")
            break
```
